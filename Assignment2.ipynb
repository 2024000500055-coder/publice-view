```python
class Car:
    def __init__(self, brand, model, year, speed=0):
        self.brand = brand
        self.model = model
        self.year = year
        self.speed = speed

    def accelerate(self, amount):
        self.speed += amount
        print(f"{self.brand} accelerated by {amount}. Current speed: {self.speed}")

    def brake(self, amount):
        self.speed -= amount
        if self.speed < 0:
            self.speed = 0
        print(f"{self.brand} slowed down. Current speed: {self.speed}")

    def display_info(self):
        print(f"Car: {self.brand} {self.model} ({self.year}) | Speed: {self.speed}")


car1 = Car("Toyota", "Corolla", 2020)
car2 = Car("BMW", "X5", 2022)

car1.accelerate(50)
car1.brake(20)
car1.display_info()

car2.accelerate(80)
car2.brake(100)
car2.display_info()
```

```python
class BankAccount:
    def __init__(self, account_holder, account_number, balance=0):
        self.account_holder = account_holder
        self.account_number = account_number
        self.balance = balance
        self.transactions = []

    def deposit(self, amount):
        self.balance += amount
        self.transactions.append(f"Deposited: {amount}")
        print(f"{amount} deposited successfully.")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Error: Insufficient balance!")
        else:
            self.balance -= amount
            self.transactions.append(f"Withdrawn: {amount}")
            print(f"{amount} withdrawn successfully.")

    def get_balance(self):
        print(f"Current Balance: {self.balance}")

    def transaction_history(self):
        print("Transaction History:")
        for t in self.transactions:
            print(t)

acc1 = BankAccount("Refat", "1001")
acc2 = BankAccount("Rahim", "1002")

acc1.deposit(1000)
acc1.withdraw(300)
acc1.deposit(500)
acc1.get_balance()
acc1.transaction_history()

print()

acc2.deposit(2000)
acc2.withdraw(500)
acc2.withdraw(200)
acc2.get_balance()
acc2.transaction_history()
```

```python
class Book:
    def __init__(self, title, author, isbn, is_available=True):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.is_available = is_available


class Library:
    def __init__(self, name):
        self.name = name
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def borrow_book(self, isbn):
        for book in self.books:
            if book.isbn == isbn and book.is_available:
                book.is_available = False
                print(f"{book.title} borrowed successfully.")
                return
        print("Book not available.")

    def return_book(self, isbn):
        for book in self.books:
            if book.isbn == isbn:
                book.is_available = True
                print(f"{book.title} returned successfully.")
                return

    def show_available_books(self):
        print("Available Books:")
        for book in self.books:
            if book.is_available:
                print(f"{book.title} by {book.author}")

lib = Library("City Library")

b1 = Book("Python Basics", "John Doe", "111")
b2 = Book("OOP Concepts", "Jane Smith", "222")
b3 = Book("Data Science", "Mike Ross", "333")
b4 = Book("AI Future", "Elon Tech", "444")

lib.add_book(b1)
lib.add_book(b2)
lib.add_book(b3)
lib.add_book(b4)

lib.borrow_book("111")
lib.borrow_book("222")
lib.return_book("111")

lib.show_available_books()
```

```python
class Employee:
    total_employees = 0

    def __init__(self, name, employee_id, department, salary):
        self.name = name
        self.employee_id = employee_id
        self.department = department
        self.salary = salary
        Employee.total_employees += 1

    def give_raise(self, percent):
        self.salary += self.salary * (percent / 100)
        print(f"{self.name}'s new salary: {self.salary}")

    def details(self):
        print(f"Name: {self.name}, ID: {self.employee_id}, Dept: {self.department}, Salary: {self.salary}")


e1 = Employee("Refat", 1, "IT", 50000)
e2 = Employee("Rahim", 2, "HR", 40000)
e3 = Employee("Karim", 3, "Finance", 45000)

e1.give_raise(10)
e2.give_raise(5)

e1.details()
e2.details()
e3.details()

print("Total Employees:", Employee.total_employees)
```

```python
class ShoppingCart:
    def __init__(self, customer_name):
        self.customer_name = customer_name
        self.items = []

    def add_item(self, name, price, quantity):
        self.items.append({"name": name, "price": price, "quantity": quantity})

    def remove_item(self, name):
        for item in self.items:
            if item["name"] == name:
                self.items.remove(item)
                print(f"{name} removed from cart.")
                return

    def total_price(self):
        total = 0
        for item in self.items:
            total += item["price"] * item["quantity"]
        return total

    def show_cart(self):
        print(f"\nShopping Cart for {self.customer_name}")
        for item in self.items:
            print(f"{item['name']} - {item['price']} x {item['quantity']}")
        print("Total:", self.total_price())

cart1 = ShoppingCart("Refat")
cart2 = ShoppingCart("Rahim")

cart1.add_item("Laptop", 800, 1)
cart1.add_item("Mouse", 20, 2)
cart1.add_item("Keyboard", 50, 1)
cart1.remove_item("Mouse")
cart1.show_cart()

cart2.add_item("Phone", 600, 1)
cart2.add_item("Headphone", 100, 1)
cart2.add_item("Charger", 25, 2)
cart2.remove_item("Charger")
cart2.show_cart()
```
